### 1c) Explain briefly why the logistic regression is not sensitive to outliers
Least squares method is sensitive to outliers, because this method calculates the squared difference between data and the mean. Outliers will distort the mean thus skewing the distribution of squared differences between data and mean. When making decision boundaries with logistic regression, on the other hand, it only takes into consideration the points that are closer to the decision boundary, so the effect of outliers on decision boundary is less. 

## 2) Logistic Regression and kNN Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = np.load('mnist_data.npy')
labels = np.load('mnist_labels.npy')
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, train_size=0.8, random_state=42)
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
print('Test accuracy: %s' %round(accuracy_score(y_test, y_pred),3))
print('Train accuracy: %s' %round(accuracy_score(y_train, y_train_pred),3))
```

```python
from sklearn.neighbors import KNeighborsClassifier
test_acc = []
train_acc = []

for k in range(1,25+2,2):
    print('When k=%s: '%k)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn_model = knn.fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)
    y_train_pred = knn_model.predict(X_train)
    
    test = accuracy_score(y_test, y_pred)
    train = accuracy_score(y_train, y_train_pred)
    test_acc.append(test)
    train_acc.append(train)
    print('Test accuracy: %s' %round(test,3))
    print('Train accuracy: %s' %round(train,3))
```

```python
xaxis = []
for k in range(1,25+2,2):
    xaxis.append(k)
    
plt.plot(xaxis, test_acc, linestyle='solid',color='blue', marker='o', markersize=4)

plt.plot(xaxis, train_acc, linestyle='solid', color='red', marker='o', markersize=4)
plt.xticks(np.arange(1, 27, 2))
plt.xlabel('k values')
plt.ylabel('accuracy score')
plt.title('Test and Train Accuracy of Different K Neighbors')
plt.legend(['Test Accuracy', 'Train Accuracy'])
```

## explain results
The plot above shows the different k values (k neighbors) and their corresponding test and train accuracy score. Both accuracy scores decrease as the k value increases. Even though small k values can give high accuracy, this would overfit the model. For example, when k=1, train accuracy will always be 1.0 (perfect) since the closest neighbor to a point is always the point itself. The optimal k value chosen based on this plot is 7. Test accuracy (blue curve) shows a steady decline as k value increases, but the decline from 5 to 7 is the smallest compared to others.

## 2f) Pros and Cons of KNN, linear regression, and logistic regression
pros of linear regression: 
- less complexity compared to other algorithms (computationally efficient)
- overfitting can be avoided with dimensionality reduction

cons of linear regression:
- sensitive to outliers
- assumes linear relationship between dependent and independent variables (this might not always hold true)
- assumes independence of variables 

pros of logistic regression: 
- makes no assumption about the distribution of classes in feature space 
- can be easily extended to multiple classes and produces probabilistic view of class predictions
- fast at classifying unkown records 

cons of logistic regression: 
- when the number of observations is less than the number of features, logistic regression would likely lead to overfitting
- can't solve non-linear problems because logistic regression has a linear decision surface. Linear separable data is rarely found in real world scenarios 
- does not work well with complex relationships 

pros of KNN: 
- versatility: can be used for classification or regression. 
- performs in non-linear situations
- performs well with features with complex relations or data with outliers

cons of KNN:
- computationally heavy
- accuracy is highly dependent on parameters such as k value and how distance is calculated 
- KNN can't handle missing values, would need to perform imputation before fitting model
